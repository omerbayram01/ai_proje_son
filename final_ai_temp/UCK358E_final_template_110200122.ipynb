```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
X_train = ss.fit_transform(X_train)
X_test = ss.transform(X_test)
lr_=LinearRegression() #linear regression model
lr_.fit(X_train, y_train) #training the model

#predicted values by rounding closest y value
y_pred_train = np.round(lr_.predict(X_train))
y_pred_test = np.round(lr_.predict(X_test))

#flattening predictions to compare easyly
y_train_r = y_train.values.reshape(-1, 1)
y_test_r = y_test.values.reshape(-1, 1)


ac_train = accuracy_score(y_train_r, y_pred_train) #accuracy of train
ac_test = accuracy_score(y_test_r, y_pred_test)

print("Train accuracy: ", ac_train)
print("Test accuracy: ", ac_test)
```

```python
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
conf_matrix_tst = confusion_matrix(y_test, y_pred_test) #test confusion matrix
conf_matrix_tr = confusion_matrix(y_train, y_pred_train) #train confusion matrix
plt.figure(figsize=(16, 6))
plt.subplot(1,2,1)
sns.heatmap(conf_matrix_tr, annot=True, fmt='d', cmap='Blues',
            xticklabels=np.unique(y_train),
            yticklabels=np.unique(y_train))
plt.title('Confusion Matrix (Train)')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.subplot(1,2,2)
sns.heatmap(conf_matrix_tst, annot=True, fmt='d', cmap='Blues',
            xticklabels=np.unique(y_test),
            yticklabels=np.unique(y_test))
plt.title('Confusion Matrix (Test)')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.tight_layout()
plt.show()
```

## Logistic Regression

```python
solvers = ['lbfgs',  'newton-cg',  'sag', 'saga']
results=pd.DataFrame(columns=["solver type","Test accuracy","Train Accuracy"])
for solver in solvers:
  logreg = LogisticRegression(multi_class='multinomial', solver=solver)
  logreg.fit(X_train, y_train)
  y_pred_test = logreg.predict(X_test)
  y_pred_train = logreg.predict(X_train)

  results.loc[len(results)] = [solver, accuracy_score(y_test, y_pred_test),accuracy_score(y_train,y_pred_train)]
```

## Random Forest

```python
max_nodes = [5, 7,11,15,19,23,27]
tr_acc=[]
test_acc=[]
for i in max_nodes:
  rndm_cls = RandomForestClassifier(n_estimators=500, max_leaf_nodes=i, random_state=42)
  rndm_cls.fit(X_train, y_train)

  y_pred_rf = rndm_cls.predict(X_test)
  y_pred_tr = rndm_cls.predict(X_train)
  tr_acc.append(accuracy_score(y_train, y_pred_tr))
  test_acc.append(accuracy_score(y_test, y_pred_rf))
```

```python
plt.plot(max_nodes,tr_acc,label="Train Accuracy")
plt.plot(max_nodes,test_acc,label="Test Accuracy")
plt.legend()
plt.xlabel("Max Leaf Nodes")
plt.ylabel("Accuracy")
plt.show()
```

By increasing the max leaf nodes we can reach higer accuracies without getting overfitting

```python
from keras.utils import to_categorical
import keras

# one hot encoding
y_test_n=y_test-1
y_train_n=y_train-1
y_tr_cat=keras.utils.to_categorical(y_train_n,num_classes=5)
y_tst_cat=keras.utils.to_categorical(y_test_n,num_classes=5)
```

```python
#creating the model
model=keras.Sequential([
    keras.Input(shape=(X_train.shape[1],)),
    keras.layers.Dense(128,activation='tanh'),
    keras.layers.Dense(64,activation='tanh'),
    keras.layers.Dense(32,activation='tanh'),
    keras.layers.Dense(5,activation='sigmoid')

])
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])


# training the model
history=model.fit(X_train, y_tr_cat, epochs=20, batch_size=32, validation_split=0.1)
train_accuracy = history.history['accuracy']
val_accuracy = history.history['val_accuracy']
# evaluation of the model on test data
loss, accuracy = model.evaluate(X_test, y_tst_cat)
print(f'Test Loss: {loss}, Test Accuracy: {accuracy}')

#plot of train and validation accuracy by epoch
plt.plot(train_accuracy,label="Train Accuracy")
plt.plot(val_accuracy,label="Validation Accuracy")
plt.legend()
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.show()
```
